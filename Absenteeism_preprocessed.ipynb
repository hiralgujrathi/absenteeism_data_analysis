# Standardize the data

```python
#preprocessing.scale
#from sklearn.preprocessing import StandardScaler
#this object will be used to scale our data i.e it will subtract the mean and divide by standardd deviation in every feature
#absenteeism_scaler=StandardScaler()
```

```python
#when we standardized the inputs we standardised the dummies as well which is bad bcz when we standardize we lose what the dummy wanted to actually interpret 
#if we could have left them in 0s and 1s we could have said for a unit change it is 7.92 times that the person will be excessively absent and a unit change in terms of dummy means disregarding this dummy to taking only this dummy in account(photo)
#so when reason_1 is given we can say it is 7.92 times that person will be absent compared to no reason given
#dummies became uninterpretable when we standardised as we lost how to compare different reasons as reasons are imp reasons
from sklearn.base import BaseEstimator,TransformerMixin
from sklearn.preprocessing import StandardScaler
class CustomScaler(BaseEstimator,TransformerMixin): 
    #this has an arguement columns that takes which columns to scale
    #NO NEED TO UNDERSTAND CODE JUST KNOW HOW TO USE IT
    def __init__(self,columns):
        self.scaler = StandardScaler()
        self.columns = columns
        self.mean_ = None
        self.var_ = None
 
    def fit(self, X, y=None):
        self.scaler.fit(X[self.columns], y)
        self.mean_ = np.array(np.mean(X[self.columns]))
        self.var_ = np.array(np.var(X[self.columns]))
        return self
 
    def transform(self, X, y=None, copy=None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), 
        columns=self.columns)
        X_not_scaled = X.loc[:,~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled, X_scaled], axis=1)[init_col_order]
```

```python
unscaled_inputs.columns.values
```

```python
#columns_to_scale=[ 'Month Value',
#       'Day of the Week', 'Transportation Expense', 'Distance to Work',
 #      'Age', 'Daily Work Load Average', 'Body Mass Index', 
 #      'Children', 'Pet']
#after removing 'Day of the Week','Daily Work Load Average','Distance to Work we get error awe defined them to standardize so to encounter it
columns_to_omit=['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4','Education']
#instead of removing manually we use list function
columns_to_scale=[x for x in unscaled_inputs.columns.values if x not in columns_to_omit]
#here list works like a loop and takes all values that are not part of columns to omit
```

```python
absenteeism_scaler=CustomScaler(columns_to_scale)
```

```python
#this line will calculate the means and stanard deviation of each feature from unscaled_inputs and this info will be stored in absenteeism scaler object so it wont be an empty object anymore as it will contain info about mean and standard deviation
#this is important because whenever we get  new data we will know that standardization info is contained in absenteeism scaler object
absenteeism_scaler.fit(unscaled_inputs),absenteeism_scaler.get_params()
```

```python
#so we have the information but the inputs are still unscaled so its just "scaling mechanism" but we need to apply it in order to get scaled inputs so we use trandform method
#so as the absenteeism scaler has mean deviation so transform peroform the calcuations and whenever we get new data we apply this method to transform it
scaled_inputs=absenteeism_scaler.transform(unscaled_inputs)
```
